## SMOTE (Synthetic Minority Over-sampling Technique)

Our dataset is imbalanced due to the higher number of 0s (non-defaults) in the `DEFAULT` column. This imbalance can cause the model to favor predicting the majority class, leading to poor performance in identifying the minority class (defaults).

#### How SMOTE Helps:
SMOTE addresses this by creating synthetic samples for the minority class, balancing the dataset. It generates new data points by interpolating between existing minority class instances and their neighbors.

#### Why SMOTE Is Important:
Without SMOTE, the model may predict the majority class more often, resulting in a high accuracy but poor predictions for defaults. By balancing the classes, SMOTE improves the model’s ability to detect defaults, enhancing overall performance.

#### Conclusion:
SMOTE helps ensure that the model learns to predict both classes effectively, improving its accuracy for the minority class and overall predictive power.

```python
# Handle Imbalanced Dataset with SMOTE
sns.countplot(x=y)#distribution of target variable before SMOTE
```

```python
# Apply SMOTE to balance the classes
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

```python
# Check the distribution after SMOTE
sns.countplot(x=y)
sns.countplot(x=y_res)
```

### Resampling Process and Data Conversion

After applying SMOTE to address class imbalance, the feature matrix (`X_res`) is converted into a NumPy array. This is typical because resampling techniques, such as SMOTE, generate synthetic data and return the result as a NumPy array.

To ensure that we retain the original column names and work with a DataFrame format, we convert `X_res` back into a pandas DataFrame (`X_resampled`). This allows us to maintain the structure of the original dataset and facilitates easier data manipulation and analysis.


```python
# Resampling and converting the data back to DataFrame
X_resampled = pd.DataFrame(X_res, columns=X.columns) #It is a dataframe cos it contsins multiple features and it is two dimensional
y_resampled = pd.Series(y_res, name='DEFAULT') #It is a series cos it is a single column of data, representing a single variable (the target) and one dimensional
```

```python
X=X.astype(int)
```

MODEL

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
# Splitting data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_resampled,y_resampled, test_size=0.3, random_state=42)
```

```python
# Standardizing the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
column_names = ['INCOME', 'SAVINGS', 'DEBT', 'R_DEBT_INCOME', 'T_EXPENDITURE_12',
       'R_EXPENDITURE_INCOME', 'R_EXPENDITURE_DEBT', 'CAT_GAMBLING',
       'CREDIT_SCORE']  

# Convert X_train and X_test to DataFrames
X_train_df = pd.DataFrame(X_train, columns=column_names)
X_test_df = pd.DataFrame(X_test, columns=column_names)

# Display both DataFrames
print("X_train:")
print(X_train_df)

print("\nX_test:")
print(X_test_df)
```

```python
y_train
```

```python
# Logistic Regression Model
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred_logistic = logistic_model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_logistic))
print("Classification Report:\n", classification_report(y_test, y_pred_logistic))
```

```python
# Random Forest Model
print("\n--- Random Forest Classifier ---")
best_params = {'max_depth': None, 'min_samples_leaf': 1, 'min_samples_split': 2, 'n_estimators': 200}

random_forest_model = RandomForestClassifier(**best_params,random_state=42)
random_forest_model.fit(X_train, y_train)
y_pred_rf = random_forest_model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
```

```python
# Decision Tree Model
print("\n--- Decision Tree Classifier ---")
decision_tree_model = DecisionTreeClassifier(random_state=42)
decision_tree_model.fit(X_train, y_train)
y_pred_dt = decision_tree_model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Classification Report:\n", classification_report(y_test, y_pred_dt))
```
